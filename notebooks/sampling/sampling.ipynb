```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ss = [0, 0.1, 1, 10, 100]
rtts = [4.010, 4.016, 4.016, 4.016, 4.041]
ths = [16.3, 16.28, 16.28, 16.09, 14.6]
```

```python
fig, ax1 = plt.subplots()
fig.set_figheight(2.4)
fig.set_figwidth(6.4)
plt.grid()

line, = ax1.plot(ss, ths, marker='o')
ax1.set_xlabel('Sampling rate (%)')
ax1.set_ylabel('Goodput (Gbps)', color=line.get_color())
ax1.tick_params(axis='y', labelcolor=line.get_color())

ax2 = ax1.twinx()
ax2.plot([], [])
line, = ax2.plot(ss, rtts, marker='v')
ax2.set_ylabel('RTT (ms)', color=line.get_color())
ax2.tick_params(axis='y', labelcolor=line.get_color())

plt.xscale('symlog')
fig.tight_layout()

plt.title('Monitoring impact')

plt.savefig('sampling.png', bbox_inches='tight')
```
